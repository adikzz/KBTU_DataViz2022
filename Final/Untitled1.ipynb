```python
bday = []
bname = []
dday = []
bplace = []
dplace = []
for i in range(len(ludi)):
    url = ludi[i]
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    
    if soup.find('span',class_='bday'):
        bname.append(url)
        bday.append(soup.find('span',class_='bday').getText())
        for j in soup.find_all('span',class_="no-wikidata"):
            if 'data-wikidata-property-id="P19"' in str(j):
                bplace.append(j.getText())
        if len(bplace)!=len(bname):
            bplace.append('null')
        if soup.find('span',class_='dday'):
            dday.append(soup.find('span',class_='dday').getText())
            for k in soup.find_all('span',class_="no-wikidata"):
                if 'data-wikidata-property-id="P20"' in str(k):
                    dplace.append(k.getText())
        else:
            dday.append('Null')
        if len(dplace)!=len(bname):
            dplace.append('null')

       


        
```
